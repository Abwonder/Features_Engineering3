    To normalize data into a range of [0,1]

 What is PCA (Principal Component Analysis)?

    A feature extraction technique that transforms data into a lower-dimensional space

    A classification algorithm used for image recognition

    A clustering algorithm used for unsupervised learning

    An outlier detection technique used for anomaly detection

Explanation: PCA (Principal Component Analysis) is a feature extraction technique that transforms data into a lower-dimensional space while preserving the most important information. It does this by finding the principal components, which are the linear combinations of the original features that explain the most variance in the data.
